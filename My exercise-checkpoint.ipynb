```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
sales = pd.read_csv('sales_data1.csv')
sales.head(10)
```

```python
sales.shape
```

```python
sales.describe()
```

```python
sales.info()
```

```python
sales['Unit_Cost'].describe()
```

```python
sales['Unit_Cost'].mean()
```

```python
sales['Unit_Cost'].median()
```

```python
sales['Unit_Cost'].plot(kind='box', vert=False, figsize=(14,6))
plt.title('Box plot')
plt.savefig('Box Plot.png', dpi = 300)
```

```python
ax = sales['Unit_Cost'].plot(kind='density', figsize=(14,6), color='red')
plt.title('Density Plot',color='red')
ax.axvline(sales['Unit_Cost'].mean(),color='green')
ax.axvline(sales['Unit_Cost'].median(),color='blue')
plt.savefig('Densty Plot.png', dpi = 300)
```

```python
ax = sales['Unit_Cost'].plot(kind='hist', figsize=(14,6), color='red')
```

```python
plt.hist(sales.Unit_Cost,color='red')
plt.title('Histograph')
plt.xlabel('Dollars')
plt.ylabel('Number of Sales')
plt.savefig('Histo Plot.png', dpi = 300)
```

```python
sales['Age_Group'].value_counts()
```

```python
sales['Age_Group'].value_counts().plot(kind='pie', figsize=(6,6))
plt.title('Pie Plot')
plt.savefig('Pie Plot.png', dpi = 300)
```

```python
ax = sales['Age_Group'].value_counts().plot(kind='bar', figsize=(14,6))
plt.ylabel('Number of Sells')
plt.savefig('Bar Plot.png', dpi = 300)
```

```python
corr = sales.corr()
corr
```

```python
fig = plt.figure(figsize=(8,8))
plt.matshow(corr, cmap='RdBu', fignum=fig.number)
plt.xticks(range(len(corr.columns)),corr.columns,rotation='vertical');
plt.yticks(range(len(corr.columns)),corr.columns);
plt.savefig('Matshow.png',dpi=300)
```
